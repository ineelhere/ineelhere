### Engineer and clean the data

```python
#merging both dataframes based on the column "title"

df = movies.merge(credits, on="title")
df.head()
```

```python
#double checking things

df.columns
```

```python
#keeping datapoints which would be of value for this project

df = df[['movie_id','title','overview','genres','keywords','cast','crew']]
df.head()
```

```python
#check for NA values

df.isnull().values.sum()
```

```python
# 3 NA values detectted. Hence removing those rows

df.dropna(inplace=True)
```

```python
#check for NA values

df.isnull().values.sum()
```

```python
#check for duplicate values

df[df.duplicated()]
```

```python
df.head()
```

```python
#function for only keeping name tags present in the columns of the dataframe

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
#calling the above function

df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
df['cast'] = df['cast'].apply(convert)
df.head()
```

```python
#function for fetching the only director name from the crew column of the dataframe

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
#calling the above function

df['crew'] = df['crew'].apply(fetch_director)
df.head()
```

```python
#remove spaces in some columns for better results in the model

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1

df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
df.head()
```

```python
#split the strings into list of words for better results and create a new field where all these will be merged.

df['overview'] = df['overview'].apply(lambda x:x.split())
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
df.head()
```

```python
#create a new dataframe with movie id, title, and tag - to be used in recommendation process

new = df.drop(columns=['overview','genres','keywords','cast','crew']) 
new.head()
```

```python
#make a whole sentence out of the list elements in tags

new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

### Building the model

```python
#transform a given text into a vector on the basis of the frequency (count) of each word 
#that occurs in the entire text

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
cv
```

```python
# calculate and transform - on the training data so that we can scale the training data 
# and also learn the scaling parameters of that data
#reference - https://datascience.stackexchange.com/questions/12321

vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
#find similarity values via cosine similarity
#reference - https://www.machinelearningplus.com/nlp/cosine-similarity/

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
#the recommender function

def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
#moment of truth

recommend("Minions")
```

```python
recommend("Jaws")
```

#### So the model recommends movie names based on other movie names, genres, actors, directors, etc!
